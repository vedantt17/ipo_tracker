```python
# Written by V
import sqlite3
import pandas as pd

conn = sqlite3.connect('../database/ipo_tracker.db')

# check row counts and null rates for every table
for table in ['ipos', 'financials', 'vc_history', 'underwriters', 'price_performance']:
    df = pd.read_sql(f'SELECT * FROM {table}', conn)
    print(f'\n=== {table} ===')
    print(f'Rows: {len(df)}')
    print('NULL rates:')
    print((df.isnull().mean() * 100).round(1).to_string())

conn.close()
```
